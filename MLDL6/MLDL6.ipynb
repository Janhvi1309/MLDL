# Task
Generate a synthetic 2D dataset suitable for clustering, apply K-Means and Hierarchical Clustering to it, visualize the results of both algorithms in a scatter plot, and summarize their performance and the resulting clusters.

## Generate Sample Data

### Subtask:
Generate a synthetic 2D dataset suitable for clustering using `make_blobs` from `sklearn.datasets`.


**Reasoning**:
First, I'll import the necessary library for generating synthetic data. Then, I'll generate a 2D dataset with specified characteristics using `make_blobs` and display the first few rows of the generated features and labels to ensure the data is created correctly.



```python
import pandas as pd
from sklearn.datasets import make_blobs

# Generate a synthetic 2D dataset
X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.8, random_state=42)

# Display the first 5 rows of the features and true labels
print("First 5 rows of features (X):\n", pd.DataFrame(X).head())
print("\nFirst 5 rows of true labels (y_true):\n", y_true[:5])
```

## Apply K-Means Clustering

### Subtask:
Apply the K-Means clustering algorithm to the generated dataset and determine cluster assignments for each data point. This step will use `sklearn.cluster.KMeans`.


**Reasoning**:
I will import the KMeans class, instantiate it with the specified parameters, fit it to the data, predict the cluster labels, and then print the first five labels to verify the clustering results.



```python
from sklearn.cluster import KMeans

# Instantiate KMeans model
kmeans = KMeans(n_clusters=4, random_state=42, n_init='auto')

# Fit the model to the data
kmeans.fit(X)

# Predict cluster labels
kmeans_labels = kmeans.predict(X)

# Print the first 5 cluster labels
print("First 5 K-Means cluster labels:", kmeans_labels[:5])
```

## Apply Hierarchical Clustering

### Subtask:
Apply Hierarchical Clustering (Agglomerative Clustering) to the same generated dataset and determine cluster assignments. This step will use `sklearn.cluster.AgglomerativeClustering`.


**Reasoning**:
I will import the `AgglomerativeClustering` class, instantiate it with the specified parameters, fit it to the data `X`, predict the cluster labels, and then print the first five labels to verify the clustering results.



```python
from sklearn.cluster import AgglomerativeClustering

# Instantiate AgglomerativeClustering model
hac = AgglomerativeClustering(n_clusters=4)

# Fit the model to the data and predict cluster labels
hac_labels = hac.fit_predict(X)

# Print the first 5 cluster labels
print("First 5 Hierarchical Clustering labels:", hac_labels[:5])
```

## Visualize Clustering Results

### Subtask:
Create a scatter plot to visualize the original data points, colored according to their cluster assignments from both K-Means and Hierarchical Clustering. Ensure proper legends are included for each cluster.


**Reasoning**:
I will import the necessary plotting libraries, create a figure with two subplots, and then plot the K-Means and Hierarchical Clustering results on each subplot respectively, ensuring proper labels, titles, and legends for better visualization and understanding.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Plot K-Means Clustering Results
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=kmeans_labels, palette='viridis', legend='full', ax=axes[0])
axes[0].set_title('K-Means Clustering Results')
axes[0].set_xlabel('Feature 1')
axes[0].set_ylabel('Feature 2')
axes[0].legend(title='Cluster')

# Plot Hierarchical Clustering Results
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=hac_labels, palette='viridis', legend='full', ax=axes[1])
axes[1].set_title('Hierarchical Clustering Results')
axes[1].set_xlabel('Feature 1')
axes[1].set_ylabel('Feature 2')
axes[1].legend(title='Cluster')

# Ensure no overlap between plot elements
plt.tight_layout()

# Display the plots
plt.show()
```
